```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function to optimize
def function_to_optimize(x):
    return (x + 3)**2

# Define the derivative (gradient) of the function
def gradient(x):
    return 2 * (x + 3)

# Hyperparameters
learning_rate = 0.1
iterations = 100
x = 2  # Starting point

# Lists to store the points for visualization
x_values = []
y_values = []

# Gradient Descent algorithm
for i in range(iterations):
    gradient_value = gradient(x)
    x = x - learning_rate * gradient_value  # Update x

    # Append the current x and y values for visualization
    x_values.append(x)
    y_values.append(function_to_optimize(x))

# Print the result
minima = function_to_optimize(x)
print(f"Local minimum at x = {x}, y = {minima}")

# Create a plot to visualize the function and optimization path
x_range = np.linspace(-10, 5, 400)
y_range = function_to_optimize(x_range)

plt.figure(figsize=(10, 6))
plt.plot(x_range, y_range, label='Function: $y = (x + 3)^2$')
plt.scatter(x_values, y_values, color='red', label='Optimization Path', marker='x')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gradient Descent Optimization')
plt.legend()
plt.grid(True)
plt.show()
```
